```python
import numpy as np
import pandas as pd

import os as os
pd.__version__
```

```python
data_path = './DataFiles/ames-housing-dataset/'
df = pd.read_csv(os.path.join(data_path,'AmesHousing.csv'),
                 engine='pyarrow', dtype_backend='pyarrow')
```

```python
df.shape
```

```python
df.head(3)
```

```python
agg_df = df.describe()
agg_df
```

```python
agg_df.iloc[0]['Order']
```

```python
missing = {}

for col in agg_df.columns:
    if agg_df.iloc[0][col]<2930.0:
        missing[col] = [round(100*(2930.0-agg_df.iloc[0][col])/2930.0,2)]

pd.DataFrame(data=missing)
```

```python
print(df.dtypes)
```

```python
agg_df2 = (df
    .select_dtypes('string')
    .describe()
)
missing = {}

for col in agg_df2.columns:
    if agg_df2.iloc[0][col]<2930.0:
        missing[col] = [round(100*(2930.0-agg_df2.iloc[0][col])/2930.0,2)]

pd.DataFrame(data=missing)
```
